### **Understanding recursion**

**Recursion in python is a programming technique where a function calls itself over and over with the objective of solving an specific problem. The function stops itself when a specific base case is met, which prevents infinite loops.**

This is its base case:

```python
def recursive_function(parameters):
    if base_case_condition:
        return base_result
    else:
        return recursive_function(modified_parameters)

## **Example 1**:

**Lets say that I want to write a function that tells us the number of steps we walk, we can approach this problem in two ways, iteratively, meaning a loop or recursively.**

````python
# Iteratively
def walk(steps):
    for step in range(1, steps + 1):
        print (f"You've taken step #{step}")
````

````python
walk(10)
````

````python
# Recursively
def walk(steps):
    print (f"You've taken step #{steps}")
    walk(steps - 1)
````

**If we execute this function, we will get a recursion error, which happens when a function calls itself too many times without meeting a stopping condition**

````python
walk(100)
````

**To prevent this, we need our base case, which is the condition that when met, it will stop the function**

````python
def walk(steps):
    if steps == 0:
        return
    walk(steps - 1)
    print (f"You've taken step #{steps}")
````

````python
walk(10)
````

## **Problem 1: Find the factorial of a number both iteratively and recursively**

````python
# Iteratively
def factorial(n):
    result = 1 # We need to keep track of a result
    if n > 0: # Factorials are generally for positive numbers in 
        for i in range(1, n + 1):
            result *= i
            n = n - 1
    return result
````

````python
range(1,5)
````

````python
factorial(5)
````

````python
# Recursively
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
````

````python
factorial(5)
````

## **Problem 2: Compute a fibonacci sequence both iteratively and recursively**

````python
# Iteratively
def fibonacci(f):
    a, b = 0, 1
    for i in range(f):
        print(a, end=" ")
        a, b = b, a + b
````

````python
fibonacci(10)
````

````python
# Recursively
def fibonacci(f):
    if f == 0:
        return 0
    elif f == 1:
        return 1
    else:
        return fibonacci(f-1) + fibonacci(f-2)
````

````python
fibonacci(12)
````
