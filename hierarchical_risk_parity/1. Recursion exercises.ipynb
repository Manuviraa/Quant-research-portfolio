## **Ejercicio 1: Recursividad con listas planas** 

**Crea una función suma_lista(lista) que tome una lista de números planos (ej. [1, 2, 3, 4]) y devuelva la suma total sin usar bucles for o while ni la función sum().**

````python
lista = [1,2,3,4,5]

def suma_lista(lista):
    if lista == []:
        return 0
    else:
        return lista[0] + suma_lista(lista[1:])
````

````python
suma_lista(lista)
````

## **Ejercicio 3: Invertir una cadena de texto (String)**
Crea invertir_texto(texto) que reciba "quant" y devuelva "tnauq". En Python, los strings se pueden dividir como las listas (texto[1:]).

````python
def invertir_texto(texto):
    if texto == "":
        return ""
    else:
        texto = texto[-1] + invertir_texto(texto[:-1])
        return texto
````

````python
invertir_texto("quant")
````

## **Ejercicio 4: Filtrar números pares**

Crea solo_pares(lista) que reciba [1, 2, 3, 4, 5] y devuelva una nueva lista solo con los pares: [2, 4].

````python
lista = [1,2,3,4,5]
````

````python
def solo_pares(lista):
    if lista == []: # Si esta vacia, regresa una lista vacia
        return []
    elif (lista[0] % 2) != 0: # Si el valor 0 NO es par, pasar la lista con los siguientes números
        return solo_pares(lista[1:])
    else:
        return [lista[0]] + solo_pares(lista[1:]) # Si el valor ES par crear una lista con ese número y sumarselo a la lista con los siguientes números
````

````python
solo_pares(lista)
````

# **Fase 3: El Puente hacia el Árbol (Anidación)**

**Para manejar listas anidadas, necesitas una herramienta que evalúe qué tipo de dato estás mirando antes de decidir si extraes el valor o si vuelves a llamar a la función. Esa herramienta es isinstance(variable, tipo)**

## **Ejercicio 5: Suma de una lista anidada**
Crea suma_anidada(lista) que reciba [1, [2, 3], 4, [[5]]] y devuelva la suma total (15).

````python
lista = [1, [2, 3], 4, [[5]]]
````

Algo que surgira cuando el problema acabe es que la lista quede vacía, si verifico la instancia en ese momento, surgira un error ya que lista[0] esta fuera del rango, por lo que primero necesito verificar si es que es una lista vacia o no lo es, si lo es, regreso un 0.
```python
if lista == []:
    return 0
```
Ahora puedo verificar la instancia, si es un int, conservare el primer valor y le sumare los valores delegados para la siguiente vez que se llame la función

```python
    elif isinstance(lista[0],int):
        return lista[0] + suma_anidada([lista[1:]])
```

Finalmente, debo de crear la condición para cuando la instancia es nuevamente una lista, como es una lista, llamo a mi función para volver a verificar las variables que esten dentro de esta con **suma_anidada(lista[0])**

También, como solo estoy verificando el índice 0, le debo de sumar los demas índices, para ello llamo a **suma_anidada(lista[1:])**

```Python
    else:
        return suma_anidada(lista[0]) + suma_anidada(lista[1:])

````python
def suma_anidada(lista):
    if lista == []:
        return 0
    elif isinstance(lista[0],int):
        return lista[0] + suma_anidada([lista[1:]])
    else:
        return suma_anidada(lista[0]) + suma_anidada(lista[1:])
````

````python
suma_anidada(lista)
````

## **Reto Final: aplanar_arbol(lista_anidada)**

**Crear una función llamada aplanar_arbol(lista_anidada) que transforme [1, [2, 3], [[4], 5]] en [1, 2, 3, 4, 5].**

Para esto debo primero crear mi caso base, en este caso es cuando no haya más números en la lista, es decir, una lista vacia. Cuando exista una lista vacia, retornare una lista vacia []

```Python
def aplanar_arbol(lista_anidada):
    return []
```

Ahora debo de planear para el siguiente caso, este sera que el valor sea realmente un número, para ello verificare su instancia. Como sera int, lo regresare en forma de lista y lo sumare a la lista de todos los números delegados

```Python
elif isinstance(lista[0],int):
    return [lista[0]] + aplanar_arbol(lista[1:])
```

Finalmente, debo de planear para el caso en el que el valor que estoy verificando NO es un número, sino una lista. En este caso debere de retornar dos cosas.
1. Mi propia función, la cuál se encargara de adentrarse en esta nueva lista
2. Mi propia función nuevamente, que se encargara de todos los demas datos delegados

```Python
else:
    return aplanar_arbol(lista[0]) + aplanar_arbol(lista[1:])

````python
lista = [1, [2, 3], [[4], 5]]
````

````python
def aplanar_arbol(lista):
    if lista == []:
        return []
    elif isinstance(lista[0],int):
        return [lista[0]] + aplanar_arbol(lista[1:])
    else:
        return aplanar_arbol(lista[0]) + aplanar_arbol(lista[1:])
````

````python
aplanar_arbol(lista)
````
