### **Notas sobre el paper "Portfolio Selection" de Harry Markowitz** 

Empezando el paper, Markowitz nos demuestra porque el únicamente buscar maximizar los retornos de un portafolio ocasiona que sea matemáticamente imposible el obtener un portafolio diversificado

***Variables:***

- $X_i$: Proporción de mi capital colocado en el activo $i$.
- $r_i$: Retorno esperado del activo $i$
- $d_i$: Tasa de descuento del activo $i$, esta es la tasa usada para traer el retorno esperado al presente


***Reglas:***

- $X_i > 0$: La proporción de mi capital colocado en un activo i siempre debera de ser mayor que 0
- $\sum X_i = 1$: La suma de el capital colocado en cada activo debera de ser siempre 100%

El retorno esperado de un portafolio de inversión es un promedio ponderado del retorno individual de cada activo que tenemos dentro de este, se calcula sumando los productos del retorno esperado de cada activo y la proporción que este activo tiene en nuestro portafolio

$$\sum X_i R_i$$

Si es que tenemos un activo que da el mayor rendimiento y nuestro objetivo es maximizar el rendimiento, entonces lo que haremos sera colocar el 100% del capital en este activo, lo que provoca que no se diversifique el portafolio

Lo siguiente que podríamos pensar es no poner todo el capital en un activo que de el mayor rendimiento, sino que podemos identificar una serie de activos que tengan un gran rendimiento comparado con la mayoría y colocar nuestro capital unicamente en estos. De esta manera se puede creer que el portafolio quedara diversificado.

Sin embargo esta asumción es erronea debido a que las acciones en el mercado no se mueven de forma independiente, sino que reaccionan a eventos, guerras, noticias y expectativas. Si es que ocurre un evento negativo lo suficientemente fuerte como una guerra, las acciones, por mayor rendimiento que tengan, caeran. Esto es conocido como **Riesgo Sistemático**

Es por ello que el inversor debe enfocarse en optimizar dos cosas:

1. Retorno esperado
2. Riesgo

## **¿Por qué funciona la diversificación?**

Markowitz plantea el siguiente proceso para poder entender porque la diversificación funciona.

Imaginemos que tenemos una variable aleatoria con $Y$, esta representara el retorno futuro de un activo y como no podemos saber con exactitud que valor tomara, esta puede tomar un número finito de valores de $y_1, y_2,\cdots,y_N$ 

La probabilidad de que $Y = y_1$ es $p_1$, $Y = y2$ es $p_2$, etc. Con ello el **valor esperado o promedio de Y sera:**

$$E = p_1y_1 + p_2y_2 + \cdots + p_Ny_N$$

Este sera el retorno que el inversor espera obtener por invertir su dinero en el activo Y

La varianza de Y sera:

$$\sigma^2 = p_1(y_1 - E)^2 + p_2(y_2 - E)^2 + \cdots + p_N(y_N - E)^2$$

### **Cálculo del retorno REAL y ESPERADO de un portafolio**

- Podemos distribuir nuestro capital $C$ como queramos en $N$ cantidad de activos. Al distribuir el capital, puede irse $X_1$ proporción de este a nuestro activo 1, $X_2$ proporción de este a nuestro activo 2, etc, tal que nuestro capital puede ser escrito como:

$$C = X_1 + X_2 + \cdots + X_N$$

- El retorno que tendran nuestros activos en el futuro no lo sabemos, por lo que podemos considerarlos como variables aleatorias $R_1, R_2, \cdots, R_N$ 

Con estos dos puntos podemos calcular el retorno real de nuestro portafolio como:

$$R = X_1R_1 + X_2R_2 + \cdots + X_NR_N$$

Es decir, es una suma ponderada de nuestras variables aleatorias

Ahora, para calcular el retorno esperado, existe una regla en probabilidad que el valor esperado de una suma ponderada es una suma ponderada de los valores esperados, por lo tanto, podemos escribir el retorno esperado $E(R)$ como:

$$E(R) = X_1 E(R_1) + X_2 E(R_2) + \cdots X_NE(R_N)$$

## **Importancia de la Varianza y Covarianza en el portafolio**

Como sabemos, la varianza es una medida de dispersión que mide que tan alejados esta un conjunto de datos de su media. Esta puede ser escrita como:

$$ \sigma^2 = \frac{\sum(x_i - \mu)^2}{N}$$

La Covarianza igual es una medida de dispersión que mide el grado de variación conjunta de dos variables

Al querer calcular la varianza de nuestro rendimiento esperado, podemos creer que bastara con solamente sumar las varianzas de cada activo, sin embargo, tenemos que recordar que las variables se mueven de forma aleatoria, por lo que algunas pueden moverse en la misma dirección o en la contraria al mismo tiempo, esto causa que su covarianza aumente o disminuya respectivamente.

Para calcularla correctamente hay que observar como es que interactua cada variable con las demas, por lo tanto, la ecuación para la varianza sera 

$$V = \sum_{i=1}^{N} \sum_{j=1}^{N} X_iX_j\sigma_{ij}$$

donde $\sigma_{ij}$ representa la covarianza entre el activo $i$ y el activo $j$

Significa que se deberan de multiplicar los pesos y las covarianzas de todos los pares posibles de activos del portafolio




# **Conjuntos eficientes y la geometría del Portafolio**

La propuesta de Markowitz es que un inversor tiene la elección de distintas combinaciones de estas dos variables ($V , E(R)$), esta  regla tiene varios nombres en la industria como riesgo beneficio, riesgo - rentabilidad y media - varianza.

Esta regla dice que el inversor debería elegir los portafolios eficientes, aquellos que tengan una varianza mínima $V$ para cierto retorno esperado $E(R)$ o el máximo retorno esperado $E(R)$ por un cierto nivel de riesgo $V$.

Es importante recalcar que  $V$ y $E(R)$ son creencias, no sabemos con certeza cuáles seran, sin embargo se usan métodos estadísticos para poder hacer una aproximación.

## **Curva Isomedia e Isovarianza**

Markowitz propone lo siguiente para poder visualizar correctamente la frontera eficiente.

Imaginemos que tenemos 3 activos $i = 1,2,3$. Para un portafolio con estos 3 activos, tenemos dos condiciones que deben cumplirse

1. $\sum_{i=1}^{3} X_i = 1$: La suma de la proporción del capital colocado en cada activo debe ser 1.
2. $X_i \geq 0$ para $i = 1,2,3$: La proporción de capital colocado en cada activo debe ser mayor o igual a 0.

Con la primera regla podemos inferir entonces que si se decide invertir tanto en el activo 1 $X_1$ y tanto en el axctivo 2 $X_2$, entonces $X_3$ puede calcularse solo:

$$X_3 = 1 - X_1 - X_2$$

De esta forma se puede trabajar en un espacio de dos dimensiones con un portafolio de 3 activos.

### **Curva Isomedia**

Una curva Isomedia se refiere a una linea que conecta a todos los portafolios que me dan el mismo retorno esperado, para poder calcular su fórmula se puede hacer lo siguiente:

Sabemos que el retorno esperado de el portafolio de 3 activos puede ser escrito como:

$$E(R) = X_1\mu_1 + X_2\mu_2 + X_3\mu_3$$

Sutituyendo $X_3$ en la fórmula obtenemos

$$E(R) = X_1\mu_1 + X_2\mu_2 + (1 - X_1 - X_2)\mu_3$$

Multiplicando

$$E(R) = X_1\mu_1 + X_2\mu_2 + \mu_3 - X_1\mu_3 - X_2\mu_3$$

Factorizando
$$E(R) = \mu_3 + X_1(\mu_1 - \mu_3) + X_2(\mu_2 - \mu_3)$$

Markowitz destaca que la forma de la ecuación es lineal ya que no contiene ninguna variable al cuadrado y que son un conjunto de líneas paralelas entre si, para comprobar esto podemos despejar nuestra ecuación de la isomedia para que tenga la forma de la ecuación de la recta y graficarlas.

En un plano, el eje X representa el peso del activo 1 $X_1$ en el portafolio, el eje Y representa el peso del activo 2  $X_2$ en el portafolio, por lo tanto debemos de despejar $X_2$ para poder graficar la curva isomedia.

1. Partimos de la ecuación del retorno esperado para un portafolio de 3 activos.

$$E(R) = \mu_3 + X_1(\mu_1 - \mu_3) + X_2(\mu_2 - \mu_3)$$

2. Pasamos del otro lado los terminos que no contienen $X_2$

$$E(R) - \mu_3 - X_1(\mu_1 - \mu_3) = X_2(\mu_2 - \mu_3)$$

3. Dividimos entre el término que multiplica a $X_2$

$$X_2 = \frac{E(R) - \mu_3}{\mu_2 - \mu_3} - \frac{\mu_1 - \mu_3}{\mu_2 - \mu_3}X_1 $$

Podemos observar como la ecuación ahora cumple con la forma de $y = mx + b$ donde:

- $y = X_2$

- $m = - \frac{\mu_1 - \mu_3}{\mu_2 - \mu_3}$

- x = $X_1$

- b = $\frac{E(R) - \mu_3}{\mu_2 - \mu_3}$

Podemos observar como la variable de retorno esperado $E(r)$ se encuentra en la intersección con el eje y, por lo que la curva isomedia en el portafolio siempre tendra la misma inclinación, independientemente de el retorno esperado que el inversionita quiera, es por ello que la curva isomedia son rectas paralelas entre si.

```python
# Librerias necesarias
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Seleccionamos 3 activos de sectores que tengan baja correlación entre si
universe = ["AAPL", "KO", "XOM"] 
inicio = "2020-01-01"
fin = "2025-12-31"

# Descargamos los precios
precios = yf.download(universe, start=inicio, end=fin, interval="1d")['Close']

# Convertimos a retornos diarios
retornos = precios.pct_change()
retornos = retornos.dropna()
```

```python
# 1. Calcular el retorno promedio historico diario de los 3 activos y convertirlo a anual
media_diaria = np.mean(retornos, axis=0)
media_anual = media_diaria * 252

#1.1 Asignamos la media anual a su activo respectivo
mu_1, mu_2, mu_3 = media_anual

# 1.2 Definimos la pendiente
m = - ((mu_1 - mu_3) / (mu_2 - mu_3))

# 1.3 Creamos un arreglo que represente la colocación de nuestro capital, este ira desde 0 a 1
x = np.linspace(0, 1, 1000)

# 1.4 Establecemos los retornos objetivo del inversionista, estos se encontraran dentro del rango del retorno historico anual promedio que obtuvimos en el paso 1
# AAPL    0.272517
# KO      0.092047
# XOM     0.189939

# 1.4 El inversionista tendra 6 rendimientos objetivo distintos que quiere alcanzar de menor a mayor
objetivos = np.linspace(0.093, 0.27, 6)
```

```python
plt.figure(figsize=(8, 6))

# 1.5 Iteramos sobre cada objetivo y calculamos su intersección con el eje x
for objetivo in objetivos:
    b = (objetivo - mu_3) / (mu_2 - mu_3)

# Calculamos la curva Isomedia 
    y = m * x + b

# La graficamos
    plt.plot(x, y, label=f'Isomedia:{objetivo * 100:.2f}%')

plt.title('Curvas Isomedia para diferentes Retornos Objetivo')
plt.xlabel("Peso en activo 1 $X_1$")
plt.ylabel("Peso en activo 2 $X_2$")
plt.grid(True, alpha=0.4) 
plt.legend()
plt.show()
```

### **Curvas Isovarianza**

Una curva de Isovarianza conecta a todos los portafolios que tienen exactamente el mismo nivel de varianza. A diferencia de las curvas isomedias, estas no son rectas, sino elipses concéntricas. Esto Markowitz lo explica desde la geometría analítica, pero puede ser explicado más eficientemente desde el algebra lineal.

Intuitivamente podemos imaginar un espacio vectorial con vectores base [1,0], [0,1] con una circunferencia con centro en el origen. A este espacio podemos aplicarle una transformación que sería una matriz de covarianza, la matriz de covarianza deformara al espacio vectorial, incluida la circunferencia. Los eigenvectores de la matriz determinan la orientación o rotación de los ejes prinicipales y los eigenvalores definen cuanto se estira o comprime la circunferencia a lo largo de estos ejes, formando asi una elipse.

Podemos observarlo de la siguiente forma:

Sabemos que la varianza de un portafolio puede escribirse como $V = w^T\Sigma w$

Donde:
- $w^T$ es el vector de los pesos de cada activo dentro del portafolio $\begin{bmatrix} X_1 \\ X_2 \\ X_3 \end{bmatrix}$
- $\Sigma$ es la matriz de covarianza

Recordando que la suma de los pesos siempre debe dar 1, sabemos que $X_3 = 1 - X_1 - X_2$, por lo tanto $X_3$ siempre esta dictado por $X_1$ y $X_2$, esto muestra que realmente no nos movemos en un espacio 3D, sino uno 2D, lo que significa que el vector de pesos $w$ esta restringido a 2 dimensiones, por lo que podemos reescribir el vector de pesos como una transformación afin a un vector 2D. Esta tendra dos partes, una constante y una variable que dependera de $X_1$ y $X_2$

$$ w = \begin{bmatrix} X_1 \\ X_2 \\ 1 - X_1 - X_2 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix} + \begin{bmatrix} X_1 \\ X_2 \\ -X_1 - X_2\end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ 1\end{bmatrix} + \begin{bmatrix} 1 & 0 \\ 0 & 1 \\ -1 & -1\end{bmatrix} \begin{bmatrix} X_1 \\ X_2\end{bmatrix}$$

A este vector podemos llamarlo $w = w_0 + Mx$, lo cuál lo sustituiremos en la ecuación original para la varianza de un portafolio

$$V =w^T \Sigma w \Longrightarrow (w_0 + Mx)^T \Sigma (w_0 + Mx) \Longrightarrow (w_0^T + x^TM^T) \Sigma (w_0 + Mx)$$
