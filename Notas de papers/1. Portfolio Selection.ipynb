### **Notas sobre el paper "Portfolio Selection" de Harry Markowitz** 

Empezando el paper, Markowitz nos demuestra matemáticamente porque el únicamente buscar maximizar los retornos de un portafolio ocasiona que sea matemáticamente imposible el obtener un portafolio diversificado

***Variables:***

- $X_i$: Proporción de mi capital colocado en el activo $i$.
- $r_i$: Retorno esperado del activo $i$
- $d_i$: Tasa de descuento del activo $i$, esta es la tasa usada para traer el retorno esperado al presente


***Reglas:***

- $X_i > 0$: La proporción de mi capital colocado en un activo i siempre debera de ser mayor que 0
- $\sum X_i = 1$: La suma de el capital colocado en cada activo debera de ser siempre 100%

El retorno esperado de un portafolio de inversión es un promedio ponderado del retorno individual de cada activo que tenemos dentro de este, se calcula sumando los productos del retorno esperado de cada activo y la proporción que este activo tiene en nuestro portafolio

$$\sum X_i R_i$$

Si es que tenemos un activo que da el mayor rendimiento y nuestro objetivo es maximizar el rendimiento, entonces lo que haremos sera colocar el 100% del capital en este activo, lo que provoca que no se diversifique el portafolio

Lo siguiente que podríamos pensar es no poner todo el capital en un activo que de el mayor rendimiento, sino que podemos identificar una serie de activos que tengan un gran rendimiento comparado con la mayoría y colocar nuestro capital unicamente en estos. De esta manera se puede creer que el portafolio quedara diversificado.

Sin embargo esta asumción es erronea debido a que las acciones en el mercado no se mueven de forma independiente, sino que reaccionan a eventos, guerras, noticias y expectativas. Si es que ocurre un evento negativo lo suficientemente fuerte como una guerra, las acciones, por mayor rendimiento que tengan, caeran. Esto es conocido como **Riesgo Sistemático**

Es por ello que el inversor debe enfocarse en optimizar dos cosas:

1. Retorno esperado
2. Riesgo

## **¿Por qué funciona la diversificación?**

Markowitz plantea el siguiente proceso para poder entender porque la diversificación funciona.

Imaginemos que tenemos una variable aleatoria con $Y$, esta representara el retorno futuro de un activo y como no podemos saber con exactitud que valor tomara, esta puede tomar un número finito de valores de $y_1, y_2,\cdots,y_N$ 

La probabilidad de que $Y = y_1$ es $p_1$, $Y = y2$ es $p_2$, etc. Con ello el **valor esperado o promedio de Y sera:**

$$E = p_1y_1 + p_2y_2 + \cdots + p_Ny_N$$

Este sera el retorno que el inversor espera obtener por invertir su dinero en el activo Y

La varianza de Y sera:

$$\sigma^2 = p_1(y_1 - E)^2 + p_2(y_2 - E)^2 + \cdots + p_N(y_N - E)^2$$

### **Cálculo del retorno REAL y ESPERADO de un portafolio**

- Podemos distribuir nuestro capital $C$ como queramos en $N$ cantidad de activos. Al distribuir el capital, puede irse $X_1$ proporción de este a nuestro activo 1, $X_2$ proporción de este a nuestro activo 2, etc, tal que nuestro capital puede ser escrito como:

$$C = X_1 + X_2 + \cdots + X_N$$

- El retorno que tendran nuestros activos en el futuro no lo sabemos, por lo que podemos considerarlos como variables aleatorias $R_1, R_2, \cdots, R_N$ 

Con estos dos puntos podemos calcular el retorno real de nuestro portafolio como:

$$R = X_1R_1 + X_2R_2 + \cdots + X_NR_N$$

Es decir, es una suma ponderada de nuestras variables aleatorias

Ahora, para calcular el retorno esperado, existe una regla en probabilidad que el valor esperado de una suma ponderada es una suma ponderada de los valores esperados, por lo tanto, podemos escribir el retorno esperado $E(R)$ como:

$$E(R) = X_1 E(R_1) + X_2 E(R_2) + \cdots X_NE(R_N)$$

## **Importancia de la Varianza y Covarianza en el portafolio**

Como sabemos, la varianza es una medida de dispersión que mide que tan alejados esta un conjunto de datos de su media. Esta puede ser escrita como:

$$ \sigma^2 = \frac{\sum(x_i - \mu)^2}{N}$$

La Covarianza igual es una medida de dispersión que mide el grado de variación conjunta de dos variables

Al querer calcular la varianza de nuestro rendimiento esperado, podemos creer que bastara con solamente sumar las varianzas de cada activo, sin embargo, tenemos que recordar que las variables se mueven de forma aleatoria, por lo que algunas pueden moverse en la misma dirección o en la contraria al mismo tiempo, esto causa que su covarianza aumente o disminuya respectivamente.

Para calcularla correctamente hay que observar como es que interactua cada variable con las demas, por lo tanto, la ecuación para la varianza sera 

$$\sum_{i=1}^{N} \sum_{j=1}^{N} X_iX_j\sigma_{ij}$$

donde $\sigma_{ij}$ representa la covarianza entre el activo $i$ y el activo $j$

Significa que se deberan de multiplicar los pesos y las covarianzas de todos los pares posibles de activos del portafolio
