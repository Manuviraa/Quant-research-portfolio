## **Estimación de Máxima Verosimilitud**

Dada una función de probabilidad a elegir, la MLE nos dice cuáles deberan de ser los parámetros para que esa función se adpate lo mejor posible a los datos que intentamos modelar

**Pasos**

1. **Elegir la PDF:** Se elige un modelo o distribución de porbabilidad como la normal, poisson, bernoulli, T-Student, etc. Esta función cuenta con parámetros desconocidos que definen su forma.

2. **Función de Verosimilitud:** El método calcula esta función la cuál representa la probabilidad de haber obtenido los datos dependiendo de los valores que tengan los parámetros

3. **Optimización:** El algoritmo busca los valores de los parámetros que maximize la probabilidad dada por la función de verosimilitud

Sea que $X_1, X_2, \cdots, X_n$ tengan una función de probabilidad conjunta, esta función puede ser definida como:

$$f(x_1, x_2, \cdots, x_n | \theta) = P(X_1 = x_1, X_2 = x_2, \cdots, X_n = x_n | \theta)$$

**Donde:** 
- $n$ son las muestras de una distribución
- $\theta$ los parámetros a estimar
- $X_n$ variables aleatorias 
- $x_n$ variables numéricas reales

Si es que las muestras son i.i.d, la función se puede simplificar multiplicando cada probabilidad individual de la siguiente manera:

Recordando que la probabilidad de que sucedan dos eventos es $P(A \cap B) = P(A) \cdot P(B|A)$, si los eventos son independientes, el saber que sucedio A no cambia nada en la probabilidad de que suceda B, por lo tanto $P(B|A) = P(B)$ por lo que $P(A \cap B) = P(A) \cdot P(B)$

Aplicando esto a nuestra función de probabilidad conjunta $$f(x_1,x_2,\cdots,x_n | \theta)$$

$$\Downarrow$$

$$P(X_1 = x_1, X_2 =x_2, \cdots, X_n = x_n| \theta)$$

$$\Downarrow$$

$$P(X_1 = x_1 | \theta) \cdot P(X_2 =x_2 | \theta), \cdots, P(X_n = x_n |  \theta)$$

$$\Downarrow$$

$$\prod_{i=1}^{n} f(x_i | \theta)$$

## **Función de Verosimilitud**

Ahora que tenemos una función de probabilidad conjunta dado un parámetro $\theta$, podemos definir una función de verosimilitud, la cuál de igual forma es una función de $\theta$.

Esta función nos dira que tan probable es obtener los datos observados dado el parámetro $\theta$, lo que queremos hacer con esta función es maximizar esa probabilidad

$$lik (\theta) = f(X_1, X_2, \cdots, X_n | \theta) = \prod_{i=1}^{n} f(X_i|\theta)$$

Para la estimación de máxima verosimilitud, lo que se hace es maximizar el logaritmo de esta función de verosimilitud

$$\hat{\theta} = \argmax lik (\theta) \Longrightarrow  \log [lik(\theta)] = \log \prod_{i=1}^{n} f(X_i | \theta) = \sum_{i=1}^{n} \log f(X_i | \theta)$$

### **Aplicandolo a la distribución de Poisson**

Como recordamos, puede ser aplicada a cualquier distribución de probabilidad, por lo que el proceso anterior se puede usar en Poisson, T-student, normal, etc. Recordando la fórmula de a distribución de Poisson $$P(X = k) = \frac{\lambda^{k} \cdot e^{-\lambda}}{k!}$$

**Donde:**
- $P(X=K)$ es la probabilidad de que ocurran K eventos
- $\lambda$ es el número de eventos promedio
- $k$ es el número de eventos


1. **Crear la función de probabilidad conjunta para Poisson**
$$f(x_1,x_2,\cdots,x_n) = \prod_{i=1}^{n} f(x_1 | \theta) = \prod_{i=1}^{n} \frac{\theta^{x_i} \cdot e^{-\theta}}{x_i!} $$

2. **Crear la función de log verosimilitud para Poisson dado un parámetro $ \theta$**

$$ \log [lik(\theta)] = \log f(X_1, X_2, \cdots X_n | \theta) = \log \prod_{i=1}^{n} \frac{\theta^{X_i} \cdot e^{-\theta}}{X_i!} = \sum_{i=1}^{n} \frac{\theta^{X_i} \cdot e^{-\theta}}{X_i}$$
